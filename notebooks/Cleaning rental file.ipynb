```python
for c in rental.select_dtypes(include='integer'):
    
    rental[c]=pd.to_numeric(rental[c], downcast='integer')
```

```python
rental.rental_date = pd.to_datetime(rental.rental_date)
rental.return_date = pd.to_datetime(rental.return_date)
```

```python
rental.info()
```

Downcasted integers and changed the type for columns rental_date and return_date to datetime 

```python
rental_dow=[]
for x in rental.rental_date:
    rental_dow.append(x.strftime('%A'))

rental.insert(2, 'rental_dow', rental_dow)
```

```python
return_dow=[]
for x in rental.return_date:
    return_dow.append(x.strftime('%A'))

rental.insert(6, 'return_dow', return_dow)
```

```python
rental.info()
```

Added a column for the name of the day of the week that the films were rented and another column for the day of the week they were returned. I would like to be able to check what days are the busiest for renting and what day people usually return the films. 

```python
rental['days_kept'] = rental['return_date'] - rental['rental_date']

rental.days_kept
```

```python
rental['days_kept']= rental['days_kept'].apply(lambda x: str(x).split()[0])
```

I also added a column that tells how many days a specific film was kept by a customer. This was done by getting the difference between return_date and rental_date and then keeping only the numbers 

```python
rental.info()
```

```python
rental.days_kept = rental.days_kept.astype(int)
rental.rental_dow = rental.rental_dow.astype('category')
rental.return_dow = rental.return_dow.astype('category')

```

```python
rental.info()
```

After adding new columns I had to change the types again. days_kept should be an integer and not an object and rental_dow and return_dow can be category since thereÂ´s only 7 days in a week
