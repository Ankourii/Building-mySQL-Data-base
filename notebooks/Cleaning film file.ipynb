```python
films.info()
```

```python
for c in films.select_dtypes(include="integer"):
    
    films[c]=pd.to_numeric(films[c], downcast="integer")
```

```python
for c in films.select_dtypes(include="float"):
    
    films[c]=pd.to_numeric(films[c], downcast="float")
```

```python
films["rating"] = films["rating"].astype("category")
```

Downcasted the types of integers and floats. For rating column which only has a few unique values I changed the type from object to category.
